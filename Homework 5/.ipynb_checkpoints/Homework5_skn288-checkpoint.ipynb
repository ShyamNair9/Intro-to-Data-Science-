### Part 1: Naive Bayes (5 Points)

1\. From your reading you know that the naive Bayes classifier works by calculating the conditional probabilities of each feature, $e_i$, occuring with each class $c$ and treating them independently. This results in the probability of a certain class occuring given a set of features, or a piece of evidence, $E$, as

$$P(c \mid E) = \frac{p(e_1 \mid c) \cdot p(e_2 \mid c) \cdot \cdot \cdot p(e_k \mid c) \cdot p(c)}{p(E)}.$$

The conditional probability of each piece of evidence occuring with a given class is given by

$$P(e_i \mid c) = \frac{\text{count}(e_i, c)}{\text{count}(c)}.$$

In the above equation $\text{count}(e_i, c)$ is the number of documents in a given class that contain feature $e_i$ and $\text{count}(c)$ is the number of documents that belong to class $c$. 

A common variation of the above is to use Laplace (sometimes called +1) smoothing. Recall the use of Laplace smoothing introduced toward the end of Chapter 3 in the section Probability Estimation. This is done in sklearn by setting `alpha=1` in the `BernoulliNB()` function (this is also the default behavior). The result of Laplace smoothing will slightly change the conditional probabilities,

$$P(e_i \mid c) = \frac{\text{count}(e_i, c) + 1}{\text{count}(c) + 2}.$$

In no more than **one paragraph**, describe why this is useful, and use the bias-variance tradeoff to justify its use. Try to think of a case when not using Laplace smoothing would result in "bad" models. Try to give an example. Be precise.

Naive Bayes works with the assumption that the attributes are mutually independent. Having this assumption may cause the classifier to be very biased. Having a high bias means that the model does not have high accuracy which is because of underfitting and high variance means that the model gives different results for small changes which is because of overfitting. The Naive Bayes classifier has high bias but low variance becuase it is a simple model using few parameters.

Now, assuming that the attributes are mutually independent of each other may not be true in every case. Assume a case where for some value of 'i', P(ei|c) is zero.  Since we are using multiplication, this would make the entire probability value for class 'c', P(c||E) to be zero. This is highly undesirable and we will get inaccurate results. For example, while filtering spams, let's say that the word 'graduate' appears just once in the spam mails and never appears in any of the ham mails. The probability P(grading| spam) will be some small value whereas it would be zero for P(grading| ham). Now, since we use muliplication to calculate probabilities, if any email consists the word 'grading' in it, it would never be classifies as 'ham' which is ridiculous. Thus, we use the techinique of smoothing. Using Laplace's smoothing ensures that we wont get the probability as zero (for any of the words in this case). This would definitely give us a more accurate estimation. Thus, introducing Laplace's smoothing helps to decrease the high bias of the Naive Bayes classifier without increasing the variance too much.

### Part 2: Text classification for sentiment analysis (20 Points)
For this part of the assignment, we are going to use a data set of movie ratings from IMDB.com. The data consists of the text of a movie review and a target variable which tells us whether the reviewer had a positive feeling towards the movie (equivalent to rating the movie between 7 and 10) or a negative feeling (rating the movie between 1 and 4). Neutral reactions are not included in the data.

The data are located in "`data/imdb.csv`". The first column is the review text; the second is the text label 'P' for positive or 'N' for negative.

1 (1 Point) \. Load the data into a pandas `DataFrame()`.

```python
import pandas as pd
data = pd.read_csv('imdb.csv', encoding='utf-8')
```

2 (1 Point)\. Code the target variable to be numeric: use the value `1` to represent 'P' and `0` to represent 'N'.

```python
data['Class'] = pd.Series(data['Class'] == "P", dtype=int)
data.head()
```

3 (2 Points)\. Put all of the text into a data frame called `X` and the target variable in a data frame called `Y`. Make a train/test split where you give 75% of the data to training. Feel free to use any function from sklearn.

```python
from sklearn.cross_validation import train_test_split

X = data['Text']
Y = data['Class']

X_train, X_test, Y_train, Y_test = train_test_split( X, Y, train_size=0.75, random_state=42)

X_train.head()
```

4 (5 Points)\. Create a binary `CountVectorizer()` and a binary `TfidfVectorizer()`. Use the original single words as well as bigrams (in the same model). Also, use an "english" stop word list. Fit these to the training data to extract a vocabulary and then transform both the train and test data. Hint - look at the API documentation for both vectorizers to see what we mean by "binary."

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords

stopwords_list = set(stopwords.words('english'))

count_vectorizer = CountVectorizer(stop_words = stopwords_list, binary = True, ngram_range=(1, 2))

tfidf_vectorizer = TfidfVectorizer(stop_words = stopwords_list, binary = True, ngram_range=(1, 2))

cv = count_vectorizer.fit(X_train)
tf = tfidf_vectorizer.fit(X_train)

cv_X_train = cv.transform(X_train)
cv_X_test = cv.transform(X_test)

tf_X_train = tf.transform(X_train)
tf_X_test = tf.transform(X_test)

cv_X_train
```

```python
for model_instance, model_name in [(LR, "LogReg"), (BNB, "Naive Bayes")]:
    for train, test, method in [(cv_X_train, cv_X_test, 'cv'), (tf_X_train, tf_X_test, 'tfidf')]:
        
        model_instance.fit(train, Y_train)
        
        fpr, tpr, thresholds = metrics.roc_curve(Y_test, model_instance.predict_proba(cv_X_test)[:,1])
        auc = metrics.roc_auc_score(Y_test, model_instance.predict_proba(test)[:, 1])
        plt.plot(fpr, tpr, label=model_name + " (" + method + ") AUC = " + str(round(auc, 4)))
        
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title("ROC Curve (AUC)")
        
plt.legend()
```

As seen from the above graph, Logistic Regression (using tfidf) model performs better than the Naive Bayes model. This happens because of the mutual independence assumption in the Naive Bayes model. As a result, Naive Bayes will give equal weight to even those features which are correlated to each other. Logistic Regression on the other hand can compensate them by giving them lower weights. Now, since we get a better plot for Logistic Regression model, it is evident that there are correlated features in our dataset. Also, the dataset is fairly large. If it was very small, then Logreg model may have overfitted it resulting in lower performance. When it comes to count vectorizer vs tfidf vectorizer, the latter performs better since the former only counts the word frequencies whereas the latter considers the word frequencies as well as how many times that word is used in the document, i.e., tfidf vectorizer gives less importance to words which are repeated several times. 

Both the Naive Bayes models perform identically. This is because we are using binary vectorizers and all the non-zero counts are converted to 1. This makes both the models essentially work the same as Bernoulli Naive Bayes model giving the same output.
